## Notebook pour l'exploration des données

### Types de donnees:

1. Données sur les produits du terroir marocain :
    * Informations sur les différents produits du terroir marocain, tels que les fruits, les légumes, les épices, l'huile d'argan, etc.
    * Caractéristiques spécifiques des produits, telles que leur origine géographique, leur méthode de production, leurs certifications (biologique, équitable, etc.).
    * Prix historiques des produits sur les marchés nationaux et internationaux.

2. Données de marché :
    * Tendances de la demande nationale et internationale pour les produits du terroir marocain.
    * Analyse des concurrents : données sur les produits similaires sur le marché et leur performance.
    * Facteurs économiques affectant la demande des produits du terroir, tels que le revenu disponible, le taux de change, etc.

3. Données démographiques et socioculturelles :
    * Profils démographiques des consommateurs nationaux et internationaux intéressés par les produits du terroir marocain.
    * Tendances des préférences des consommateurs en matière de produits alimentaires, y compris les préférences pour les produits locaux et biologiques.

4. Données saisonnières et climatiques :
    * Effets des saisons sur la disponibilité et la demande des produits du terroir.
    * Conditions météorologiques et climatiques affectant la production et la qualité des produits du terroir marocain.

5. Données réglementaires et commerciales :
    * Règlements nationaux et internationaux sur l'importation et l'exportation des produits alimentaires.
    * Accords commerciaux et barrières tarifaires affectant le commerce des produits du terroir marocain.

6. Données de vente et de distribution :
    * Canaux de distribution utilisés pour commercialiser les produits du terroir marocain, tels que les supermarchés, les marchés locaux, les boutiques en ligne, etc.
    * Ventes historiques des produits du terroir à travers différents canaux de distribution.

## Importation des bibliothèques

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Exploration des données

```python
# Set up paths
EXPORTS_BY_PRODUCT_V_PATH = r"../data/raw/List_of_products_exported_by_Morocco_V.csv"
EXPORTS_BY_PRODUCT_GIV_PATH = r"../data/raw/List_of_products_exported_by_Morocco_GIV.csv"
EXPORTS_BY_PRODUCT_SIVP_PATH = r"../data/raw/List_of_products_exported_by_Morocco_SIVP.csv"


# Load data
exports_by_product_v = pd.read_csv(EXPORTS_BY_PRODUCT_V_PATH)
exports_by_product_giv = pd.read_csv(EXPORTS_BY_PRODUCT_GIV_PATH)
exports_by_product_sivp = pd.read_csv(EXPORTS_BY_PRODUCT_SIVP_PATH)
```

```python
# Visualize exports by product (values)
exports_by_product_v.head()
```

```python
# Visualize exports by product (growth in value)
exports_by_product_giv.head()
```

```python
# Visualize exports by product (share in value percentage)
exports_by_product_sivp.head()
```

```python
# List columns of exports by product (values)
print("Exports by product (values) columns: ", exports_by_product_v.columns)
print("Exports by product (growth in value) columns: ", exports_by_product_giv.columns)
print("Exports by product (share in value percentage) columns: ", exports_by_product_sivp.columns)
```

```python
# Rename exports by product columns
years = np.arange(2003, 2023)
exports_by_product_v.columns = ['code', 'product'] + [f'value_{year}' for year in years]
exports_by_product_giv.columns = ['code', 'product'] + [f'growth_{year}' for year in years[:-1]] + ['value_2022']
exports_by_product_sivp.columns = ['code', 'product'] + [f'share_{year}' for year in years]
```

```python
# Merge exports by product values, growth in value and share in value percentage
exports_by_product_v = exports_by_product_v.drop(columns=['code'])
exports_by_product_giv = exports_by_product_giv.drop(columns=['code', 'value_2022'])
exports_by_product_sivp = exports_by_product_sivp.drop(columns=['code'])
exports_by_product = pd.merge(exports_by_product_v, exports_by_product_sivp, on='product')
exports_by_product = pd.merge(exports_by_product, exports_by_product_giv, on='product')
exports_by_product.head()
```

```python
# Visualize exports by product
plt.figure(figsize=(10, 5))
sns.barplot(x='product', y='value_2022', data=exports_by_product[:10]) # first 10 values
plt.xticks(rotation=90)
plt.title('Exports by product in 2022')
plt.show()
```

```python
# Save data to interim
exports_by_product.to_csv(r"../data/interim/exports_by_product.csv", index=False)
```
